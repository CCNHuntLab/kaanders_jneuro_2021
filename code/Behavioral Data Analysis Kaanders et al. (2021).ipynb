### Figure 1C

```python
# make left - right rank difference variable for attribute trials
main_data_attribute['L_R_rankdiff']=0
for row_id, row in enumerate(main_data_attribute.values):
    if main_data_attribute['Side_First_Cue'].iloc[row_id]==1:
        main_data_attribute['L_R_rankdiff'].iloc[row_id]=main_data_attribute['Rank1'].iloc[row_id]-main_data_attribute['Rank2'].iloc[row_id]
    else:
        main_data_attribute['L_R_rankdiff'].iloc[row_id]=main_data_attribute['Rank2'].iloc[row_id]-main_data_attribute['Rank1'].iloc[row_id]
```

```python
# make dummy variable describing whether participant sampled at least one additional cue
main_data_attribute['Third_Sample']=[1 if x>0 else 0 for x in main_data_attribute['NumMoves']]
```

```python
main_data_attribute.head()
```

```python
# plot psychometric functions predicting choice and sampling
subjects=list(range(1,31))
sns.set_style('whitegrid', {'axes.linewidth': 1, 'axes.edgecolor':'black'})
fig=figure(figsize=(12,8))
def model(x):
    y = 1 / (1 + np.exp(-x))
    return y
#run the classifier
clf = sklearn.linear_model.LogisticRegression(C=1e5)
logit1 = []
logit2 = []
plot = fig.add_subplot(1,1,1)
# get means and SEs for all possible value differences
for x in subjects:
    mean_prob_choice=[]
    var_prob_choice = []
    mean_prob_sample = []
    var_prob_sample = []
    rankdiffs=range(-4,5)
    for y in rankdiffs:
        mean_prob_choice.append(mean(main_data_attribute[main_data_attribute['L_R_rankdiff']==y]['Left']))
        var_prob_choice.append(scipy.stats.sem(main_data_attribute[main_data_attribute['L_R_rankdiff']==y]['Left']))
        mean_prob_sample.append(mean(main_data_attribute[main_data_attribute['L_R_rankdiff']==y]['Third_Sample']))
        var_prob_sample.append(scipy.stats.sem(main_data_attribute[main_data_attribute['L_R_rankdiff']==y]['Third_Sample']))
X_test = np.linspace(-5,5,300)
# get fit predicting choice
clf.fit(main_data_attribute['L_R_rankdiff'][:, np.newaxis], main_data_attribute['Left'])
logit1 = model(X_test*clf.coef_).ravel()
# get fit predicting third sample
coefs = np.polyfit(main_data_attribute['L_R_rankdiff'].values.flatten(),  main_data_attribute['Third_Sample'].values.flatten(), 3)
# plot fits
plot.plot(X_test, logit1, zorder=1, linewidth=3, color='#831E29')
plot.plot(X_test, np.polyval(coefs, X_test), linewidth=3, color="#45B1C8")
plot.set_ylabel('Probability', size=30)
plot.set_xlabel('Left - right picture \n Rank Difference', size=30)
plot.tick_params(axis='both', which='major', labelsize=20)
for y in rankdiffs:
    plot.errorbar(rankdiffs, mean_prob_choice, yerr=var_prob_choice, fmt = 'o', marker='o',markerfacecolor='black', markeredgecolor = 'black', ms = 7, ecolor='gray')
    plot.errorbar(rankdiffs, mean_prob_sample, yerr=var_prob_sample, fmt = 'o', marker='o',markerfacecolor='lightgray', markeredgecolor = 'black', ms = 7, ecolor='gray')
plot.set_ylim(0, 1.01)
plot.set_xlim(-5, 5.5)
plot.grid(True)
grid(False)

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='#831E29', lw=5),
                Line2D([0], [0], color="#45B1C8", lw=5)]

plot.legend(custom_lines, ['Prob(Left)', 'Prob(Sample)'], prop={'size': 24}, bbox_to_anchor=(1.05, 1), loc='upper left')

sns.despine(right=True, top=True)
plt.tight_layout()
plt.subplots_adjust(top=0.88, left=0.15)

fig.suptitle('Attribute Trials', fontsize = 32)
```

### T-test: Choice predicted by value difference (Attribute Trials)

```python
# run logistic regression predicting choice from value difference for each subject separately
new=[]
Choice_coefs=pd.DataFrame(data=[new])
Choice_coefs['Coef']=0
Choice_coefs['P']=0
main_data_attribute['Constant']=1

for subj in main_data['Participant'].unique():
    print('%s' % subj)
    cols = ['Constant','L_R_rankdiff']
    X=main_data_attribute[main_data_attribute['Participant']==subj][cols]
    y=main_data_attribute[main_data_attribute['Participant']==subj]['Left']
    logit_model=sm.Logit(y,X)
    result=logit_model.fit()
    Choice_coefs.loc[subj-1,'Coef']=result.params[1]
    Choice_coefs.loc[subj-1,'P']=result.pvalues[1]
```

```python
# run t-test across coefficients
scipy.stats.ttest_1samp(Choice_coefs['Coef'],0)
```

### T-test: Sampling predicted by value difference (Attribute Trials)

```python
# get absolute value difference
main_data_attribute['abs_rankdiff']=abs(main_data_attribute['L_R_rankdiff'])
```

```python
# run logistic regression predicting additional information sampling from value difference for each subject separately
new=[]
Choice_coefs=pd.DataFrame(data=[new])
Choice_coefs['Coef']=0
Choice_coefs['P']=0

for subj in main_data['Participant'].unique():
    print('%s' % subj)
    cols = ['Constant','abs_rankdiff']
    X=main_data_attribute[main_data_attribute['Participant']==subj][cols]
    y=main_data_attribute[main_data_attribute['Participant']==subj]['Third_Sample']
    logit_model=sm.Logit(y,X)
    result=logit_model.fit()
    Choice_coefs.loc[subj-1,'Coef']=result.params[1]
    Choice_coefs.loc[subj-1,'P']=result.pvalues[1]
```

```python
scipy.stats.ttest_1samp(Choice_coefs['Coef'],0)
```

### Figure 1D

```python
# compute difference in value for option trials
main_data_option['Sum']=main_data_option['Rank1']+main_data_option['Rank2']
# subtract mean rank
main_data_option['Rank_Sum']=main_data_option['Sum']-6
```

```python
# make variable indicating if revealed option was chosen
main_data_option['Choice_First_Option']=0
for row_id, row in enumerate(main_data_option.values):
    if main_data_option['Choice'].iloc[row_id]==main_data_option['Side_First_Cue'].iloc[row_id]:
        main_data_option['Choice_First_Option'].iloc[row_id]=1
```

```python
# make variable indicating whether participant sampled additional cues
main_data_option['Third_Sample']=[1 if x>0 else 0 for x in main_data_option['NumMoves']]
```

```python
sns.set(font_scale=2)
sns.set_style('whitegrid', {'axes.linewidth': 1, 'axes.edgecolor':'black'})
fig=figure(figsize=(12,8))
def model(x):
    y = 1 / (1 + np.exp(-x))
    return y
#run the classifier
clf = sklearn.linear_model.LogisticRegression(C=1e5)
logit1 = []
logit2 = []
plot = fig.add_subplot(1,1,1)
#get mean and SE for each possible rank difference
for x in subjects:
    mean_prob_choice=[]
    var_prob_choice=[]
    mean_prob_sample=[]
    var_prob_sample=[]
    rankdiffs=range(-4,5)
    for y in rankdiffs:
        mean_prob_choice.append(mean(main_data_option[main_data_option['Rank_Sum']==y]['Choice_First_Option']))
        var_prob_choice.append(scipy.stats.sem(main_data_option[main_data_option['Rank_Sum']==y]['Choice_First_Option']))
        mean_prob_sample.append(mean(main_data_option[main_data_option['Rank_Sum']==y]['Third_Sample']))
        var_prob_sample.append(scipy.stats.sem(main_data_option[main_data_option['Rank_Sum']==y]['Third_Sample']))
X_test = np.linspace(-5,5,300)
# get fit for choice as a function of value difference
clf.fit(main_data_option['Rank_Sum'][:, np.newaxis], main_data_option['Choice_First_Option'])
logit1 = model(X_test*clf.coef_).ravel()
# get fit for sampling as a function of value difference
coefs = np.polyfit(main_data_option['Rank_Sum'].values.flatten(),  main_data_option['Third_Sample'].values.flatten(), 3)
#plot fits
plot.plot(X_test, logit1, zorder=1, linewidth=3, color='#831E29')
plot.plot(X_test, np.polyval(coefs, X_test), linewidth=3, color="#45B1C8")
for y in rankdiffs:
    plot.errorbar(rankdiffs, mean_prob_choice, yerr=var_prob_choice, fmt = 'o', marker='o',markerfacecolor='black', markeredgecolor = 'black', ms = 7, ecolor='gray')
    plot.errorbar(rankdiffs, mean_prob_sample, yerr=var_prob_sample, fmt = 'o', marker='o',markerfacecolor='lightgray', markeredgecolor = 'black', ms = 7, ecolor='gray')

plot.set_ylabel('Probability', size=30)
plot.set_xlabel('First + Second picture \n Sum - Mean', size=30)
plot.tick_params(axis='both', which='major', labelsize=20)

plot.set_ylim(0, 1.01)
plot.set_xlim(-5, 5)
plot.grid(True)
grid(False)

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='#831E29', lw=5),
                Line2D([0], [0], color="#45B1C8", lw=5)]

plot.legend(custom_lines, ['Prob(Choice=1)', 'Prob(Sample)'], prop={'size': 24}, bbox_to_anchor=(1.05, 1), loc='upper left')


sns.despine(right=True, top=True)
plt.tight_layout()
plt.subplots_adjust(top=0.88, left=0.15)
fig.suptitle('Option Trials', fontsize = 32)
```
